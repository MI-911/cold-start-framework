```python
import json
import matplotlib.pyplot as plt

data = json.load(open('label_cov.json', 'r'))

for label, pairs in data.items():
    x = [pair[0] for pair in pairs]
    y = [pair[1] * 100 for pair in pairs]
    
    plt.scatter(x, y, label=label)

plt.title('Information from 1-length interview')
plt.xlabel('Popularity rank')
plt.ylabel('NDCG@10')
plt.legend()

plt.savefig('pop_gain.pdf', bbox_inches='tight')
plt.show()
```
